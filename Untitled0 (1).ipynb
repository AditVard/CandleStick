```python
import pandas as pd
import os

file_paths = ["/content/46900pe.csv","/content/47100ce.csv","/content/47400ce.csv","/content/47400pe.csv","/content/47600ce.csv","/content/47800pe.csv","/content/47200ce.csv"]
output_folder = "/content/processed_files"
os.makedirs(output_folder, exist_ok=True) #to create a folder
for file_path in file_paths:
  df = pd.read_csv(file_path)
  df['date'] = pd.to_datetime(df['date'])
  df_filtered = df[df['date'].dt.date == pd.Timestamp("2024-01-10").date()]
  df_filtered.head(5)
  df_filtered.set_index('date', inplace=True)
  df_5minutes = df_filtered.resample('5T').agg({
      'open': 'first',
      'high': 'max',
      'low': 'min',
      'close': 'last',

  }).dropna()
  output_file = os.path.join(output_folder, os.path.basename(file_path).replace(".csv", "_5min.csv"))
  df_5minutes.to_csv(output_file)
```

```python
import pandas as pd

file_paths = [
    "/content/46900pe.csv",
    "/content/47100ce.csv",
    "/content/47400ce.csv",
    "/content/47400pe.csv",
    "/content/47600ce.csv",
    "/content/47800pe.csv","/content/47200ce.csv"
]

jan10_data_dict = {}  # Store data for plotting

for file_path in file_paths:
    data_frame = pd.read_csv(file_path)
    data_frame['date'] = pd.to_datetime(data_frame['date'])

    jan10_data = data_frame[data_frame['date'].dt.date == pd.Timestamp("2024-01-10").date()]
    if jan10_data.empty:
        print(f"No data for 10th Jan in {file_path}")
        continue

    High = jan10_data['high'].max()
    Low = jan10_data['low'].min()
    Close = jan10_data['close'].iloc[-1]

    # Get the time at which the highest value occurred (without using idxmax)
    high_row = jan10_data[jan10_data['high'] == High]
    high_time = high_row['date'].values[0]

    P = (High + Low + Close) / 3
    R1 = P + 0.382 * (High - Low)
    R2 = P + 0.618 * (High - Low)
    R3 = P + 1.382 * (High - Low)
    S1 = P - 0.382 * (High - Low)
    S2 = P - 0.618 * (High - Low)
    S3 = P - (High - Low)

    print(f"\nFibonacci Pivot Points for {file_path}:")
    print(f"High: {High:.2f} (Occurred at: {high_time})")
    print(f"Low: {Low:.2f}, Close: {Close:.2f}")
    print(f"Pivot Point (P): {P:.2f}")
    print(f"R1: {R1:.2f}, R2: {R2:.2f}, R3: {R3:.2f}")
    print(f"S1: {S1:.2f}, S2: {S2:.2f}, S3: {S3:.2f}\n")

    jan10_data_dict[file_path] = jan10_data

```

```python
#This is extra I added to help undertsand each hour our closing price

import matplotlib.pyplot as plt
import seaborn as sns

for file_path, jan10_data in jan10_data_dict.items():
    jan10_data_hourly = jan10_data.resample('H', on='date').last()  # Get last close per hour
    plt.figure(figsize=(10, 5))
    sns.lineplot(x='date', y='close', data=jan10_data_hourly)
    plt.xticks(rotation=45)
    plt.title(f"Hourly Closing Price on 10th Jan - {file_path.split('/')[-1]}")
    plt.show()

```
